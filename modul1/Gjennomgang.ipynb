n√• vil jeg finne ut hvor mye hver feauture betyr for modellen, det finnes flere m√•ter √• gj√∏re dette p√•, jeg skal n√• g√• gjennom 3 forskjellige type: Feature importance, a more reliable technique is permutation importance, Partial Dependence Plots

1. Random forests can provide a measure called feature importance which reveals the significance of each feature in making predictions.
2. The idea behind permutation importance is that if a feature is important for the predictions of a trained model, then a random shuffling of the data in the corresponding column should drastically reduce the model's accuracy. On the other hand, a random shuffling of an unimportant feature (for a given trained model) shouldn't impact the model much.
3. The idea is to plot predictions while increasing a specified feature. For example, what happens to the predicted probability of diabetes as the glucose level increases? As this is an important feature, we expect it to change quite a bit (in fact, since we're talking about diabetes, we expect the probability to increase). And what about unimportant features like age?

```python
#feature importance:
importances = rf.feature_importances_
importances
#Hvis en feature ofte brukes til √• dele opp data p√• en m√•te som gj√∏r gruppene ‚Äúrenere‚Äù,
#f√•r den h√∏y importance-verdi.
#VIKTIG: Hvis to features er sterkt korrelert (f.eks. h√∏yde og vekt),
#kan modellen bruke bare den ene av dem ‚Äî og da f√•r den andre lav importance,
#selv om den egentlig inneholder samme informasjon.
#feature importance kan man ikke stole blindt p√•, derfor er det bedre √• bruke permutation importance
```

```python
indices = np.argsort(importances)[::-1]

for f in range(X.shape[1]): 
    print(f'{X.columns[indices[f]]}: {np.round(importances[indices[f]],2)}')
```

```python
# Set the figure size and the number of decimal places to round the importances to
fig_size = (12, 8)
decimals = 2

# Create the horizontal bar plot of feature importances
f, ax = plt.subplots(figsize=fig_size)
plt.barh(X.columns[indices], np.round(importances[indices], decimals))
ax.set_xlabel("Relative importance")
ax.set_title("Feature importances")
plt.show()
```

```python
#permutation importance:
```

```python
# Import the permutation_importance function
from sklearn.inspection import permutation_importance

# Calculate feature importances using permutation importance
n_repeats = 30
random_state = 0
r = permutation_importance(rf, X_test, y_test,
                            n_repeats=n_repeats,
                            random_state=random_state)
```

```python
for i in r.importances_mean.argsort()[::-1]:
    print(f"{X_test.columns[i]:<8}"
    f"{r.importances_mean[i]:.3f}"
    f" +/- {r.importances_std[i]:.3f}")
```

```python
from sklearn.inspection import PartialDependenceDisplay
import matplotlib.pyplot as plt

# Set the figure size
fig_size = (10, 6)

# Create a PDP for "sepal length (cm)" for class 0
f, ax = plt.subplots(figsize=fig_size)
PartialDependenceDisplay.from_estimator(
    rf, X_train, ['sepal length (cm)'],
    kind='both',
    subsample=50,
    ax=ax,
    target=0
)

plt.title('Partial dependence for "sepal length (cm)" (class 0: Setosa)')
plt.show()
```

```python
def f(x):
    return np.sin(2*np.pi*x)
```

```python
X_exact = np.arange(0.0, 1.0, 0.01)
plt.plot(X_exact, f(X_exact), 'g')
plt.title("Exact connection between x and t, t=sin(2*pi*x)")
plt.show()
```

```python
def generate_data(N, sigma, seed=42):
    """
    Generate N data points sampled uniformly at random, 
    with noise added to the labels (sampled from the 
    normal distribution with standard deviation sigma). 
    Seed set to 42 by default.
    Return: X, y
    """
    np.random.seed(seed)
    
    # Random values between 0 and 1 sampled uniformly
    X = np.random.rand(N)
    
    # Corresponding values of T, with some added noise
    y = f(X) + sigma*np.random.randn(N)
    
    return X, y
```

```python
N_train = 20
sigma=0.2

X_train, y_train = generate_data(N=N_train, sigma=sigma)

plt.plot(X_train, y_train, 'bo')
plt.xlabel('X')
plt.ylabel('y')
plt.title("Training data")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,6))

ax.plot(X_exact, f(X_exact), 'r--', linewidth=2, 
        label='Exact values')
ax.plot(X_train, y_train, 'bo', markersize=8, 
        fillstyle='none', label='Training data')

plt.xlabel('X')
plt.title(f'''Plot of a training set of N = {N_train} points. 
          The green curve is the function used to generate the data''', fontsize=16)

ax.legend()
ax.grid()
plt.show()
```

```python
# More advanced techniques: Bayesian optimization
from hyperopt import hp, fmin, tpe, Trials, STATUS_OK, space_eval
```

```python
from sklearn.model_selection import cross_val_score

def objective(params):
    # Our objective is to find the hyperparameters that gives the lowest mean squared error 
    # on validation data
    
    print(f"Using parameters {params}")
    # Our model
    model = RandomForestRegressor(**params, n_jobs=-1, random_state=42)
    # The max cross-val score with current paramaters:
    score = cross_val_score(model, X_train, y_train, cv=3, scoring="neg_mean_squared_error").mean()
    # We want to minimize the mean absolute error loss, not its negative:
    loss = -score
    return {'loss': loss, 'status': STATUS_OK}
```

```python
param_space = {
    'n_estimators': hp.choice('n_estimators', [10, 50, 100, 150, 200]),
    'max_depth': hp.choice('max_depth', [5, 50, None]),
    'max_features': hp.choice('max_features', [2, 3, None])
}
```

```python
import hyperopt.pyll.stochastic
```

```python
print(hyperopt.pyll.stochastic.sample(param_space))
```

```python
# By using the trials object we can check what goes on in each trial:
trials = Trials()
```

```python
best = fmin(fn=objective, space=param_space, algo=tpe.suggest, 
            max_evals=8, trials=trials)
```

```python
best
```

```python
space_eval(param_space, best)
```

```python
model = RandomForestRegressor(**space_eval(param_space, best), random_state=42, n_jobs=-1)
```

```python
model.fit(X_train, y_train)
mean_squared_error(y_test, model.predict(X_test))
```
