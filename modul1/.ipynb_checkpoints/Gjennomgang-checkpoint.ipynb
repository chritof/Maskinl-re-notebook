```python
pd.plotting.scatter_matrix(wine_df[features].iloc[:, :5], c=wine_df['label'], figsize=(15,15), marker='o', s=60)
plt.show()
```

```python
import seaborn as sns
sns.pairplot(wine_df[features[5:10] + ['label']], hue='label', palette="colorblind")
plt.show()
```

```python
sns.set_theme(style="whitegrid")
sns.lineplot(data=wine_df.iloc[:, :3], palette="tab10", linewidth=2.5)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(wine_df)

X_scaled_df = pd.DataFrame(X_scaled, columns=wine_df.columns)

sns.lineplot(data=X_scaled_df.iloc[:, :3], palette="tab10", linewidth=2.5)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X = wine_dataset['data'][:, :12]
y = wine_dataset['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=42, n_estimators=1000) # random_state for reproducibility
```

```python
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
rf.score(X_test, y_test)
```
