```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import sklearn
```

```python
# To make the notebook reproducible
seed = 42
np.random.seed(seed)
```

```python
import warnings
warnings.simplefilter(action='ignore', category=DeprecationWarning)
```

```python
np.random.seed(seed)
m = 20
X = 3 * np.random.rand(m, 1)
y = 1 + 0.5 * X + np.random.randn(m, 1) / 1.5
```

```python
plt.plot(X,y, 'b.')
plt.axis([0, 3, 0, 4])
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)

X_new = np.array([[0], [4]])
y_pred = lin_reg.predict(X_new)
```

```python
plt.plot(X,y, 'b.')
plt.plot(X_new, y_pred, 'r')
plt.axis([0, 3, 0, 4])
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
```

```python
def get_polyreg(degree):
    poly_reg = make_pipeline(PolynomialFeatures(degree=degree, include_bias=False),
                            StandardScaler(), 
                             LinearRegression())
    
    return poly_reg
```

```python
def plot_poly_regression(degree):
    poly_reg = get_polyreg(degree)
    poly_reg.fit(X, y)
    X_new = np.linspace(0, 4, 100).reshape(100, 1)
    y_pred = poly_reg.predict(X_new)
    plt.plot(X,y, 'b.')
    plt.plot(X_new, y_pred, 'r-')
    plt.axis([0, 3, 0, 4])
    plt.show()
    return poly_reg
```

```python
poly_reg_2 = plot_poly_regression(2)
```

```python
poly_reg_10 = plot_poly_regression(10)
```

```python
poly_reg_15 = plot_poly_regression(15)
```

```python
#begynner med ridge regression:
from sklearn.linear_model import Ridge

def get_ridgereg(degree, alpha):
    ridge_reg = make_pipeline(PolynomialFeatures(degree=degree),
                              StandardScaler(), 
                              Ridge(alpha=alpha))
    
    return ridge_reg
```

```python
def plot_ridge_reg(degree, alpha):
    ridge_reg = get_ridgereg(degree, alpha)
    ridge_reg.fit(X, y)
    X_new = np.linspace(0, 4, 100).reshape(100, 1)
    y_pred = ridge_reg.predict(X_new)
    plt.plot(X,y, 'b.')
    plt.plot(X_new, y_pred, 'r-')
    plt.axis([0, 3, 0, 4])
    plt.show()
    return ridge_reg
```

```python
_ = plot_poly_regression(degree=10)
```

```python
_ = plot_ridge_reg(10, 0.001)
```

```python
# kan n prve Lasso regression
from sklearn.linear_model import Lasso

def get_lassoreg(degree, alpha):
    lasso_reg = make_pipeline(PolynomialFeatures(degree=degree),
                              StandardScaler(), 
                              Lasso(alpha=alpha))
    
    return lasso_reg
```

```python
def plot_lasso_reg(degree, alpha):
    lasso_reg = get_lassoreg(degree, alpha)
    lasso_reg.fit(X, y)
    X_new = np.linspace(0, 4, 100).reshape(100, 1)
    y_pred = lasso_reg.predict(X_new)
    plt.plot(X,y, 'b.')
    plt.plot(X_new, y_pred, 'r-')
    plt.axis([0, 3, 0, 4])
    plt.show()
    return lasso_reg
```

```python
_ = plot_lasso_reg(15, 0.01)
```

```python
#prver Elastic net:
from sklearn.linear_model import ElasticNet
def get_elasticnet(degree, alpha, l1_ratio=0.5):
    elastic_net = make_pipeline(PolynomialFeatures(degree=degree),
                              StandardScaler(), 
                              ElasticNet(alpha=alpha, l1_ratio=l1_ratio))
    
    return elastic_net
```

```python
def plot_elasticnet(degree, alpha):
    elastic_net = get_elasticnet(degree, alpha)
    elastic_net.fit(X, y)
    X_new = np.linspace(0, 4, 100).reshape(100, 1)
    y_pred = elastic_net.predict(X_new)
    plt.plot(X,y, 'b.')
    plt.plot(X_new, y_pred, 'r-')
    plt.axis([0, 3, 0, 4])
    plt.show()
    return elastic_net
```

```python
_ = plot_elasticnet(15, 0.01)
```
